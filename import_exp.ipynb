```python
def initial_population_generator(population_size):
    out_list = []
    for i in range(population_size):
        out_list.append([random.randint(1,20),random.randint(1,20),random.randint(10,200),random.randint(10,200)])
    return out_list
```

```python
def add_avg_loss(list_of_parameters):
    out_list = []
    for i in list_of_parameters:
        avg_loss = ai_stuff.big_function(i[0],i[1],i[2],i[3])
        out_list.append((i,avg_loss))
    return out_list
```

```python
def sort_by_loss(list_of_parameters_with_losses):
    out_list = []
    list_of_losses = []
    
    for i in list_of_parameters_with_losses:
        list_of_losses.append(i[1])
    list_of_losses.sort()
    for i in list_of_losses:
        for j in list_of_parameters_with_losses:
            if i == j[1]:
                out_list.append(j)
    return out_list
```

```python
def binary_reproduce(solution_list_1,solution_list_2,slice_index):
    child_list_1 = solution_list_1[:slice_index] + solution_list_2[slice_index:]
    child_list_2 = solution_list_2[:slice_index] + solution_list_1[slice_index:]
    return (child_list_1,child_list_2)
```
